```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

def get_top_viral_songs():
    # Set up Spotify credentials
    client_id = 'b084cdc702e24c6b8e09e2b9525c0c73'
    client_secret = 'b07bba8c198d4d639298f4cba7b1b1c9'

    # Set up the Spotify client credentials manager
    client_credentials_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
    sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)

    # Get the top 25 viral tracks
    playlist_id = '37i9dQZEVXbKuaTI1Z1Afx'  # Viral 50 playlist ID
    results = sp.playlist_tracks(playlist_id, limit=25)

    # Print the top 25 viral songs
    for idx, item in enumerate(results['items']):
        track = item['track']
        print(f"{idx+1}. {track['name']} - {', '.join([artist['name'] for artist in track['artists']])}")

# Call the function to retrieve and print the top 25 viral songs
get_top_viral_songs()
```
