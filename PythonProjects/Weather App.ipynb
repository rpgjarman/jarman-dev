**Weather App Project**

```python
import requests

def celsius_to_fahrenheit(temp_celsius):
    """Converts temperature from Celsius to Fahrenheit."""
    return round((temp_celsius * 9/5) + 32)

def get_weather(city):
    """Fetches and displays weather information for a given city."""
    api_key = "d02ca1acefac324de5190828b910d7bf"
    # Replace with your OpenWeatherMap API key
    url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"

    response = requests.get(url)
    data = response.json()

    if data.get("cod") == "404":
        print("City not found.")
    else:
        weather = data.get("weather")
        if weather and len(weather) > 0:
            main_info = weather[0].get("main")
            description = weather[0].get("description")
            temp_celsius = data["main"].get("temp")
            humidity = data["main"].get("humidity")

            if main_info and description and temp_celsius and humidity:
                temp_fahrenheit = celsius_to_fahrenheit(temp_celsius)

                # Display weather information
                print(f"Weather in {city}")
                print(f"Main: {main_info}")
                print(f"Description: {description}")
                print(f"Temperature: {temp_fahrenheit}°F")
                print(f"Humidity: {humidity}%")
            else:
                print("Weather data incomplete or unavailable.")
        else:
            print("Weather data not available.")

# Prompt user to enter a city
city_name = input("Enter the name of a city: ")

# Call the function to fetch and display weather information
get_weather(city_name)
```
