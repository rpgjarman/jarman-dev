**Task 3**: Create a function to find a random word from the list
- import the `random` module
- Define a function called `find_word() that takes two parameters:
  - `num_letters`: an integer representing the length of the word to find
  - `words`: the list of words to pick a random word from
- Use the `random` function `choice()` to pick a random word from the `words`. Save it as `word`
- Create a while loop that will continue while the length of `word` is not equal to the parameter `num_letters`. Inside the loop, reset the value of `word` to a random word in `words`
- return the random word that is the correct length once one is found
- Test your function by calling the `find_word()` function with your saved word list from the cell above and providing at least 3 different numbers for the number of letters. Print out the result after each call.

````
import random
def find_word(num_letters, words):
  word = random.choice(words)
  while len(word) != num_letters:
    word = random.choice(words)
  return word
result1 = find_word(2, words_list)
print(result1)
result1 = find_word(10, words_list)
print(result1)
result1 = find_word(5, words_list)
print(result1)
````

**Task 4**: Create a function to get the number of letters for the word
- Define a function called `get_number_of_letters()` that takes no parameters
- Create an infinite loop. Inside the loop
  - Get input from the user by prompting them to enter the number of letters the word for the game should have (1 - 14 letters)
  - Check if the input value is a digit. If it is, cast it as an integer. If not, tell the user that their entry is invalid and that they should enter a number between 1 and 14
  - If the input value is a digit, check if the number is between 1 and 14 inclusive. If it is, break the loop. If not, tell the user that their entry is invalid and that they should enter a number between 1 and 14
- When the loop is complete, return the number of letters the word should be as an integer.
- Test your function by calling the `get_number_of_letters()` function, trying various inputs, and printing the result

````
def get_number_of_letters():
  while True:
    user_in = input("Enter a number of letters between 1 and 14: ")
    if user_in.isdigit():
      user_in = int(user_in)
      if 1 <= user_in <= 14:
        break
      else:
        print("Your entry is invalid")
    else:
      print("Your entry is invalid")
  return user_in
result2 = get_number_of_letters()
print(result2)
````

**Task 6**: Create a function to make the board
- Define a function called `build_board()` that takes one parameter:
  - `length` the size of the board, which is he same as the length of the secret word
- Create a variable called `board` and initialize it to the empty list
- Use a for loop with `range()` iteration to loop as many times as the length of the board. Each time through the loop, append a `-` to the board list.
- After the loop is complete, return the board list
- Test your function by calling it and printing the results. You should have a list that is the same length as the `length` provided containing dashes `-` in each element

````
def build_board(length):
  board = []
  for item in range(length):
    board.append("-")
  return board
result4 = build_board(len(result1))
print(result4)
````

**Task 7**: Create a function to determine if a letter has been guessed
- Define a function called `check_letter()` that takes two parameters:
  - `letter` the letter to check for being guessed
  - `guessed` a list of previously guessed letters
- Check if the lowercase version of the letter is in the guessed list. If it is, return `False`. Otherwise return `True`

````
def check_letter(letter, guessed):
  if letter.lower() in guessed:
    return False
  else:
    return True
````

  `Incorrect Guesses: <number> Guessed letters: <letters that have been guessed>`
